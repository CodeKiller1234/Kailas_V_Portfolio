```python
#we have convert the images to a numeric datas 
flat_data=[]             #  a array to store the numeric values of the images
target = []              #each image are set as dog op cat
categories = ['Cat','Dog']
datadir='/content/drive/MyDrive/datasets/Dog_cats'
```

```python
import numpy as np

for i in categories:
  print('loading....Category:')
  path=os.path.join(datadir,i) #datadir path , join i---i has cat intially  #/content/drive/MyDrive/datasets/Dog_cats/Cats this path
  for  img in os.listdir(path):
    img_array=imread(os.path.join(path,img)) ##/content/drive/MyDrive/datasets/Dog_cats/Cats + 1st image in Cat ,ie it gives the first image path ,image is in array
    
    img_resized=resize(img_array,(150,150,3)) 
    flat_data.append(img_resized.flatten())    #flatten is used for getting in one dim
    target.append(categories.index(i))         #index of i -- position of cat ie,0
  print("loaded Category :",i,"Succesfully")
    
```

```python
df=pd.DataFrame(flat_data)
df
```

```python
df['target']=target
```

```python
df
```

```python
X=df.iloc[:,:-1].values
y=df.iloc[:,-1].values
```

```python
from sklearn import svm
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=.2,random_state=77)
```

```python
clf=svm.SVC()
clf.fit(X_train,y_train)
```

```python
y_pred=clf.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score,ConfusionMatrixDisplay
print('accuracy',accuracy_score(y_test,y_pred)*100)
print(ConfusionMatrixDisplay.from_predictions(y_test,y_pred))
```
